```python
# Q1
import logging as lg
lg.basicConfig(filename="error.log",level=lg.INFO)
def fib(n):
    try:
        if n==0 or n==1:
            return n
        else:
            return fib(n-1)+fib(n-2)
    except Exception as e:
        lg.info(e)

try:
    n=int(input())
    print(fib(n))
except Exception as e:
    lg.info(e)
```

```python
# Q2
import logging as lg
lg.basicConfig(filename="error.log",level=lg.INFO)
def fact(n):
    try:
        if n<=1:
            return n
        else:
            return n*fact(n-1)
    except Exception as e:
        lg.info(e)

try:
    n=int(input())
    print(fact(n))
except Exception as e:
    lg.info(e)
```
