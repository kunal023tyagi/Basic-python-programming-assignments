```python
# Q1
import logging as lg
lg.basicConfig(filename='error.log' , level=lg.INFO)
def fun(n,length):
    try:
        li=[]
        for i in range(n):
            li.append(input("Enter the string: "))
        for i in li:
            if len(i)>length:
                print(i)
    except Exception as e:
        lg.info(str(e))
try:
    n = int(input("Enter length of list: "))
    length = int(input("Enter length: "))
    fun(n,length)
except Exception as e:
    lg.info(str(e))
```

```python
# Q2
import logging as lg
lg.basicConfig(filename='error.log' , level=lg.INFO)
def remove_char(string , i):
    try:
        a = string[:i-1]
        b = string[i:]
        return a+b
    except Exception as e:
        lg.info(str(e))
try:
    n = input("Enter string: ")
    char = int(input("Enter ith character to remove: "))
    print(remove_char(n,char))
except Exception as e:
    lg.info(str(e))
```

```python
# Q3
def split_string(string):
    list_string = string.split(' ')    
    return list_string

def join_string(list_string):
    string = '-'.join(list_string)    
    return string

try:
    string = input("Enter the string: ")
    print(split_string(string))
    print(join_string(string))
except Exception as e:
    lg.info(str(e))
```

```python
# Q4
def check_binary_string(string):
    try:
        temp = '01'
        flag = 1
        for i in string:
            if i not in temp:
                flag=0
                break
            else:
                continue
        if flag:
            print("Binary string")
        else:
            print("Not a Binary string")
    except Exception as e:
        lg.info(str(e))
try:
    string = input("Enter the string: ")
    check_binary_string(string)
except Exception as e:
    lg.info(str(e))
```

```python
# Q5
def uncommon_words(string1 , string2):
    my_string=''
    for i in string1:
        if i not in string2:
            my_string+=i+" "
    for i in string2:
        if i not in string1:
            my_string+=i+" "
    print(my_string)
try:
    string1 = input("enter string1: ")
    string2 = input("enter string2: ")
    uncommon_words(string1.split(' ') , string2.split(' '))
except Exception as e:
    lg.info(str(e))
```

```python
# Q6
def duplicate_characters(string):
    try:
        duplicate=[]
        for char in string:
            if string.count(char) > 1:
                duplicate.append(char)
        print(set(duplicate))
    except Exception as e:
        lg.info(str(e))
try:
    string = input("enter string: ")
    duplicate_characters(string)
except Exception as e:
    lg.info(str(e))
```
