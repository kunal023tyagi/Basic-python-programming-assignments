```python
import logging as lg
lg.basicConfig(filename="error.log",level=lg.INFO)
def sum_array(n):
    try:
        li = []
        sum=0
        for i in range(1,n+1):
            num  = int(input("Enter {} number:".format(i)))
            li.append(num)

        for i in li:
            sum+=i

        return sum
    except Exception as e:
        lg.info(e)

try:
    n=int(input("Enter the size of array: "))
    print("Sum = " , sum_array(n))
except Exception as e:
    lg.info(e)
```

```python
# Q2
import logging as lg
lg.basicConfig(filename="error.log",level=lg.INFO)
def Max_Array(n):
    try:
        li = []
        for i in range(1,n+1):
            num  = int(input("Enter {} number:".format(i)))
            li.append(num)
        return max(li)
    except Exception as e:
        lg.info(e)

try:
    n=int(input("Enter the size of array: "))
    print("Max in array is = " , Max_Array(n))
except Exception as e:
    lg.info(e)
```

```python
# Q3
import logging as lg
lg.basicConfig(filename="error.log",level=lg.INFO)
def Array_Rotation(n):
    try:
        li = []
        for i in range(1,n+1):
            num  = int(input("Enter {} number:".format(i)))
            li.append(num)
            
        j = len(li)-1
        for i in range(len(li)//2):
            temp = li[i]
            li[i] = li[j]
            li[j] = temp
            j-=1
        print(li)
    except Exception as e:
        lg.info(e)

try:
    n=int(input("Enter the size of array: "))
    Array_Rotation(n)
except Exception as e:
    lg.info(e)
```

```python
#Q5
import logging as lg
lg.basicConfig(filename="error.log",level=lg.INFO)
def Monotonic(n):
    try:
        li = []
        for i in range(1,n+1):
            num  = int(input("Enter {} number:".format(i)))
            li.append(num)
            
        temp=0
        result=True
        for i in range(len(li)-1):
            if(li[i]>=li[i+1]):
                if(temp==1):
                    result=False
                    break
                else:
                    temp=0
                    continue
            elif(li[i]<=li[i+1]):
                if(i!=0 and temp==0):
                    result=False
                    break
                else:
                    temp=1
                    continue
        return result
    except Exception as e:
            lg.info(e)
try:
    n=int(input("Enter the size of array: "))
    if(Monotonic(n)==True):
        print("Monotonic Array")
    else:
        print("Not a Monotonic Array")
except Exception as e:
    lg.info(e)
```
