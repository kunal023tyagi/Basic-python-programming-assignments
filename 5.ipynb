```python
# Q2
import logging as lg
lg.basicConfig(filename="error.log",level=lg.INFO)
class HCF:
    def __init__(self,num1,num2):
        self.num1=num1
        self.num2=num2
        
    def my_fun(self):
        try:
            if self.num1 < self.num2:
                smaller = self.num1
            else:
                smaller = self.num2

            for i in range(1,smaller+1):
                if((self.num1%i==0) and (self.num2%i==0)):
                    hcf=i
            return hcf
        except Exception as e:
            lg.info(str(e))
            
    def __str__(self):
        return "LCM class created"

try:
    num1 = int(input("Enter 1st number: "))
    num2 = int(input("Enter 2nd number: "))
    obj = HCF(num1 , num2)
    print("HCF of {} and {} is {}".format(num1 , num2 , obj.my_fun()))
except Exception as e:
    lg.info(str(e))
```

```python
# Q3
import logging as lg
lg.basicConfig(filename="error.log",level=lg.INFO)
class Conversion:
    def __init__(self,num):
        self.num=num
        
    def DecimalToBinary(self):
        try:
            return bin(self.num)
        except Exception as e:
            lg.info(str(e))
    def DecimalToOct(self):
        try:
            return oct(self.num)
        except Exception as e:
            lg.info(str(e))
            
    def DecimalToHex(self):
        try:
            return hex(self.num)
        except Exception as e:
            lg.info(str(e))
            
    def __str__(self):
        return "Conversion class created"

try:
    num = int(input("Enter number: "))
    obj = Conversion(num)
    print("Binary of {} is {} Octal of {} is {} Hex of {} is {}".format(num,obj.DecimalToBinary(),num,obj.DecimalToOct(),num,obj.DecimalToHex()))
except Exception as e:
    lg.info(str(e))
```

```python
# Q4
import logging as lg
lg.basicConfig(filename="error.log",level=lg.INFO)
class ASCII:
    def __init__(self,num):
        self.num=num
        
    def mainFun(self):
        try:
            return ord(self.num)
        except Exception as e:
            lg.info(str(e))
            
    def __str__(self):
        return "ASCII class created"

try:
    num = input("Enter character: ")
    obj = ASCII(num)
    print("ASCII value of {} is {}".format(num , obj.mainFun()))
except Exception as e:
    lg.info(str(e))
```

```python
# Q5
import logging as lg
lg.basicConfig(filename="error.log",level=lg.INFO)
class Calculator:
    
    def __init__(self,num1,num2):
        self.num1 = num1
        self.num2= num2
        
    def Addition(self):
        try:
            return self.num1+self.num2
        except Exception as e:
            lg.info(str(e))
    
       
    def Subtraction(self):
        try:
            if self.num1 > self.num2:
                return self.num1-self.num2
            else:
                return self.num2-self.num1
        except Exception as e:
            lg.info(str(e))

    def Multiplication(self):
        try:
            return self.num1*self.num2
        except Exception as e:
            lg.info(str(e))
       
    def Division(self):
        try:
            if self.num1 > self.num2:
                return self.num1//self.num2
            else:
                return self.num2//self.num1
        except Exception as e:
            lg.info(str(e))
    def __str__(self):
        return "Calculator class created"

try:
    num1 = int(input("Enter Num1: "))
    num2 = int(input("Enter Num2: "))
    obj = Calculator(num1,num2)
    print(obj.Addition())
    print(obj.Subtraction())
    print(obj.Multiplication())
    print(obj.Division())
except Exception as e:
    lg.info(str(e))
```
